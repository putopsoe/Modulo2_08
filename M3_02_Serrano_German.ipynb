**1) Código a evaluar:**

```python
#Ejercicio1

def safe_divide(numerator, denominator):
    """Divide numerator entre denominator manejando errores concretos.

    Devuelve None si hay un error y muestra un mensaje explicativo.
    """
    try:
        resultado = numerator / denominator
    except ZeroDivisionError:
        print("Error: división por cero. No se puede dividir entre 0. Por favor, proporciona un divisor distinto de 0.")
        return None
    except TypeError:
        print("Error: tipos inválidos para la división. Asegúrate de pasar números (int o float).")
        return None
    else:
        return resultado

print("Prueba en libreta - caso enunciado: 7 / 0")
print("Resultado:", safe_divide(7, 0))
print()
print("Prueba en libreta - caso válido: 10 / 2")
print("Resultado:", safe_divide(10, 2))
print()
print("Prueba en libreta - tipo inválido: '2' / 10")
print("Resultado:", safe_divide('2', 10))
```

```python
#Ejercicio4

def safe_add(a, b):
    """Intenta sumar/concatenar a y b manejando errores de tipo.

    Estrategia:
    - Intentar la operación directa (a + b).
    - Si falla por TypeError, intentar convertir a números (int) y sumar.
    - Si eso falla, convertir ambos a cadenas y concatenar.
    - Si ninguna conversión es posible, informar al usuario.
    """
    try:
        return a + b
    except TypeError:
        try:
            ra = int(a)
            rb = int(b)
            print("Nota: convertí los operandos a enteros y realicé la suma.")
            return ra + rb
        except Exception:
            try:
                print("Nota: convertí los operandos a cadenas y realicé la concatenación.")
                return str(a) + str(b)
            except Exception:
                print("Error: tipos incompatibles. Convierte los operandos a números o cadenas según lo que quieras hacer.")
                return None

resultado = safe_add("2", 10)
print("Resultado:", resultado)
```
